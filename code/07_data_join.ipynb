# **データ結合**

7.1 データの読み込み

```python
import pandas as pd
```

```python
# データの読み込み
transaction_history = pd.read_csv('transaction_history.csv', dtype={'customer_id': str})  # 顧客の購入履歴データ
product_master = pd.read_csv('product_master.csv')  # 商品マスターのデータ
```

7.2 複数のデータを結合

```python
# transaction_historyとProduct_masterをproduct_idで結合（Left Join)し、marged_dataオブジェクトに格納
merged_data = transaction_history.merge(product_master, on='product_id', how='left')

# 結果の表示
merged_data.head()
```

7.3 データの基本情報の確認

```python
# marged_dataの行数と列数を確認
merged_data.shape
```

7.4 Right Join

```python
# Transaction_historyとProduct_masterをproduct_idで結合（Right JOIN)し、marged_data2オブジェクトに格納
merged_data2 = transaction_history.merge(product_master, on='product_id', how='right')

# 結果の表示
merged_data2.head()
```

```python
# marged_data2の行数と列数を確認
merged_data2.shape
```

7.6 Inner Join

```python
# transaction_historyとProduct_masterをproduct_idで結合（innner join)し、marged_data3オブジェクトに格納
merged_data3 = transaction_history.merge(product_master, on='product_id', how='inner')

# 結果の表示
merged_data3.head()
```

```python
# marged_data3の行数と列数を確認
merged_data3.shape
```

```python
# customer_idとdestination_city列を抽出
extracted_data = merged_data3[['customer_id', 'destination_city']]

# 結果の表示
extracted_data.head()
```

7.9 顧客別の訪れた年と回数の集計

```python
# extracted_dataのcustomer_id別に、destination_Cityごとの出現回数をカウント（顧客別に、どの都市に何度訪れたかのクロス集計表）
# customer_idごとにdestination_cityの出現回数をカウントしdestination_countsオブジェクトに格納
destination_counts = extracted_data.groupby('customer_id')['destination_city'].value_counts().unstack(fill_value=0)

# 結果の表示
destination_counts.head()
```

**欠損値補完のための結合**

7.10 アンケートデータの読み込み

```python
# 顧客のアンケートデータををquestionary_dataというデータフレームに読み込み
import pandas as pd
questionary_data = pd.read_csv('questionary_data.csv', dtype={'customer_id': str})

# 結果の表示
questionary_data.head()
```

```python
# 欠損値の確認
questionary_data.isnull().sum()
```

7.12 基本統計量の算出

```python
# questionary_dataオブジェクトのannual_income    の基本統計量を算出
questionary_data['annual_income'].describe()
```

7.13 annual_incomeの欠損値を平均値で補完

```python
#qestionary_dataオブジェクトのannual_incomeの欠損値のセルに対し、annual_incomeの平均値を代入
#その後、代入したデータセットを作成し、qestionary_data2として格納し,表示する

# annual_incomeの平均値を計算
mean_annual_income = questionary_data['annual_income'].mean()

# annual_incomeの欠損値を平均値で補完
questionary_data2 = questionary_data.fillna({'annual_income': mean_annual_income})

# 結果を表示
questionary_data2.head()
```

```python
# 確認のため、qestionary_data2の欠損値を出力
questionary_data2.isnull().sum()
```
