11.18 グレースケール変換

```python
img_1_gray = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
#https://note.nkmk.me/python-opencv-bgr-rgb-cvtcolor/
cv2_imshow(img_1_gray)
```

```python
img_2_gray = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_2_gray)
```

11.19 配列操作を用いた色変換

```python
# BGRからRGBへ変換（方法①：チャンネル順を逆にするだけ）
# 高速だが単純なスライスであり、色空間の意識はない
img_2_RGB_1 = img_2[:, :, ::-1]
cv2_imshow(img_2_RGB_1)

# BGRからRGBへ変換（方法②：OpenCV公式のカラースペース変換）
# より信頼性が高く、特殊な画像形式にも対応可能
img_2_RGB_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
cv2_imshow(img_2_RGB_2)
```

11.22 上下反転

```python
#元画像
cv2_imshow(img_1)

#上下反転
img_1_flip_ud = cv2.flip(img_1, 0)
cv2_imshow(img_1_flip_ud)
cv2.imwrite(path + '/save.jpeg', img_1_flip_ud)
```

11.23 左右反転（上下左右反転）

```python
#元画像
cv2_imshow(img_2)

#左右反転
img_2_flip_lr = cv2.flip(img_2, 1)
cv2_imshow(img_2_flip_lr)

#上下左右反転
#img_2_flip_ud_lr = cv2.flip(img_2, -1)
#cv2_imshow(img_2_flip_ud_lr)
```

11.24 ぼかし

```python
img_1_blur_1 = cv2.blur(src=img_1, ksize=(1, 1))
img_1_blur_2 = cv2.blur(src=img_1, ksize=(5, 5))
img_1_blur_3 = cv2.blur(src=img_1, ksize=(10, 10))

# 1枚ずつ表示
cv2_imshow(img_1)
cv2_imshow(img_1_blur_1)
cv2_imshow(img_1_blur_2)
cv2_imshow(img_1_blur_3)
```
