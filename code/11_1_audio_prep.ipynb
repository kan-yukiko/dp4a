11.11 音声ファイルの分割

```python
# import speech_recognition as sr
from pydub import AudioSegment
from pydub.silence import split_on_silence
```

```python
# ファイルを読み込み
sound = AudioSegment.from_file("sample.wav", format="wav")

# 音源を短く分割
chunks = split_on_silence(
    sound,

    # 2000ms以上の無音がある箇所で分割
    min_silence_len=2000,

    # -40dBFS以下で無音とみなす
    silence_thresh=-40,

    # 分割後600msだけ、無音を残す
    keep_silence=600
)

# 分割数の表示
print(len(chunks))
```

11.12 分割したファイルの格納

```python
#分割したデータを特定のフォルダに格納
%mkdir /content/分割
```

```python
# 前で分割した音源をフォルダに保存
FileNum_list = list(range(len(chunks)))
for i, chunk in enumerate(chunks):
    chunk.export(u"/content/分割/"+str(FileNum_list[i])+".wav", format="wav")
```

```python
import whisper

# Whisperモデルの読み込み（base, small, medium, largeなど指定。大きくなるほど処理が重くなります）
model = whisper.load_model("small")

# m4aファイルを読み込んで文字起こし
result = model.transcribe("sample.wav")

# 結果を表示
print(result["text"])
```

11.14 ノイズの除去

```python
!pip install librosa noisereduce soundfile numpy
import librosa
import noisereduce as nr
import soundfile as sf

# 音声ファイルの読み込み
y, sr = librosa.load("sample.wav")

# ノイズリダクション
reduced_noise = nr.reduce_noise(y=y, sr=sr)

# 処理後の音声をreduced_noise.wavというファイルに保存
sf.write("reduced_noise.wav", reduced_noise, sr)
```
