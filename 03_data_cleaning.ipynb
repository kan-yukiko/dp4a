データの欠損を確認

3.14 読み込んだデータの欠損を確認するisnull()メソッド

```python
#読み込んだデータの欠損を確認
df.isnull()
```

```python
# ① 各列ごとのNull（欠損値）をカウント
null_counts_per_column = df.isnull().sum()
print("各列の欠損値数：")
print(null_counts_per_column)

# ② 全列がNullの行を抽出して、その件数をカウント
all_null_rows_count = df[df.isnull().all(axis=1)].shape[0]

# ③ 結果の出力
print(f"\n全ての列が欠損している行数: {all_null_rows_count}")
```

```python
# すべての列が欠損している行のみを抽出
all_null_df = df[df.isnull().all(axis=1)]

# 行番号（インデックス）を含めて出力
print("全ての列が欠損している行：")
print(all_null_df.reset_index().rename(columns={"index": "行番号"}))
```
